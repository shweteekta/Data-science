```python
raw_content = rq.get("https://en.wikipedia.org/wiki/Special:Random")  #Special:Random is use to get the random article
content = raw_content.text
title_start = content.find("<title>")
title_end = content.find("</title>")
title_raw = content[title_start+len('<title>'):title_end-len(' - Wikipedia')] 
title = title_raw.strip(" - Wikipedia")   # Strip is use to remove the content
```

```python

img = Image.open('image.png')
draw = ImageDraw.Draw(img)
draw.text((70,135),title,(255,255,255),font=font)
img.save('text_image.png')
IPythonImage(filename='text_image.png')
```

```python
import requests as rq
```

```python
raw_content = rq.get("https://www.amazon.in/All-new-Echo-Dot-3rd-Gen/dp/B0792KTHKK/ref=sr_1_1?s=amazon-devices&ie=UTF8&qid=1553324306&sr=1-1&keywords=echo+dot")  #Special:Random is use to get the random article
```

```python
content = raw_content.text
print(content)
```

```python
title_start = content.find("<title>")
```

```python
title_end = content.find("</title>")
```

```python

x=content[title_start+len('<title>'):title_end]
print(x)
```

# Dynamic Approach for printing only title

```python
end = x.find(':')
print(x[0:end])
```

```python
print(content[title_start+len('<title>'):title_end-len(': Amazon.in: Kindle Store')])
```

```python
price_start = content.find('id="priceblock_saleprice"')
```

```python
print(content[price_start+108:price_start+116])
```

```python
star = content.find('class="a-icon-alt"')
print(star)
```

```python
starend = content.find("</span")
print(starend)
```

```python
print(content[star+19:star+37])
```

```python
start = content.find('id="productTitle"')
raw1 = content[start+len('id="productTitle" class="a-size-large">'):]
```

```python
end = raw1.find('</span')
print(raw1[:end].strip())
```
